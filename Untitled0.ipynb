```python
import glob
train_files = []
mask_files = glob.glob('/content/kaggle_3m/*/*_mask*') #모든 mask

for i in mask_files:
    train_files.append(i.replace('_mask','')) #sort보다 _mask해서 묶는게 더 정확함
```

```python
len(train_files) #파일 갯수
```

```python
import matplotlib.pyplot as plt#밑에 코드 쓰기위해 import
import cv2

rows,cols=3,3 #여기서 부터 얻어온것
fig=plt.figure(figsize=(10,10))
for i in range(1,rows*cols+1):
    fig.add_subplot(rows,cols,i)
    img_path=train_files[i]
    msk_path=mask_files[i]
    img=cv2.imread(img_path)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    msk=cv2.imread(msk_path)
    plt.imshow(img)
    plt.imshow(msk,alpha=0.4)
plt.show()
```

```python
import numpy as np #Image_Segmentation에서 가져옴
from PIL import Image

img_ds = []
mask_ds = []
for img, mask in zip(train_files[:3000], mask_files[:3000]):  #3926개지만 다운될수있어서 조금 적게 3000개까지 한다는 의미
  img_arr = np.array(Image.open(img).resize((128,128))) #폴더안의 모든경로를 배열로 바꿈, #원래 256이지만 밑에 코드 돌리다가 다운돼서 작게 만들어줌
  mask_arr = np.array(Image.open(mask).resize((128,128)))
  img_ds.append(img_arr)
  mask_ds.append(mask_arr)
img_ds = np.array(img_ds)
mask_ds = np.array(mask_ds) 
mask_ds = np.expand_dims(mask_ds,3)
img_ds.shape, mask_ds.shape
```

```python
np.max(img_ds),np.max(mask_ds) #보통 mask 갯수는 적은데 max가 255인거봐서 0아니면 255인거같아서 밑에 표그려봄
```

```python
img_ds=img_ds/255.0 #정규화
mask_ds=mask_ds/255.0
```

```python
plt.subplot(121)
plt.hist(img_ds[0].flatten())
plt.subplot(122)
plt.hist(mask_ds[0].flatten()) #마스크 0또는1
```
