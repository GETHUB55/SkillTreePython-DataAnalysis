```python
import glob
train_files = []
mask_files = glob.glob('/content/kaggle_3m/*/*_mask*')

for i in mask_files:
    train_files.append(i.replace('_mask',''))
```

```python
len(train_files)
```

```python
import matplotlib.pyplot as plt
import cv2

rows,cols=3,3
fig=plt.figure(figsize=(10,10))
for i in range(1,rows*cols+1):
    fig.add_subplot(rows,cols,i)
    img_path=train_files[i]
    msk_path=mask_files[i]
    img=cv2.imread(img_path)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    msk=cv2.imread(msk_path)
    plt.imshow(img)
    plt.imshow(msk,alpha=0.4)
plt.show()
```

```python
import numpy as np
from PIL import Image

img_ds = []
mask_ds = []
for img, mask in zip(train_files[:3000], mask_files[:3000]):  #폴더안의 모든경로를 배열로 바꿈
  img_arr = np.array(Image.open(img).resize((128,128)))
  mask_arr = np.array(Image.open(mask).resize((128,128)))
  img_ds.append(img_arr)
  mask_ds.append(mask_arr)
img_ds = np.array(img_ds)
mask_ds = np.array(mask_ds) 
mask_ds = np.expand_dims(mask_ds,3)
img_ds.shape, mask_ds.shape
```

```python
np.max(img_ds),np.max(mask_ds)
```

```python
img_ds=img_ds/255.0
mask_ds=mask_ds/255.0
```

```python
plt.subplot(121)
plt.hist(img_ds[0].flatten())
plt.subplot(122)
plt.hist(mask_ds[0].flatten())
```
